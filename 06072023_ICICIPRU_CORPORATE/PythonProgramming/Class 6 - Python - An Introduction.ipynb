# Data Visualization

**Data Visualization gives us the faster way of understanding the data**

**Exploratory Data Analysis (EDA)**

    1) Univariate Analysis - Where we are taking only one variable and analysing that, for example whats the data distribution of my customers based on the age, bank balance (Numerical). We can create following charts like Histograms, Box Plot, these two charts will give understanding about the numerical columns distribution. For categorical column we have the count plot where we can count lets say how many males and females are there (Gender) 
    
    2) Bivariate Analysis - Where we are taking two variables and analysing that, it can be numerical and numerical or it can be numerical and categorical. Numerical and Numerical example: Area of the house increasing then house price is also increasing. Here we can create the scatter plot. In this example its a positive relationship. In the other example, as the distance from the city is increasing the house price is decreasing. Its a negative relationship. Categorical and Numerical Example: How much sales by gender, here we can create a bar chart, pie chart
    
    3) Multivariate Analysis: Multivariate analysis is when there are three or more than three variables for analysis. It can be all numericals, It can be one numerical and rest categoricals, it can be one categorical and rest numericals. For example: House Price, Area of the house, House type. From the tips data we were having Gender, Smoker and tips.

# Matplotlib

**Using the bar function we are getting the vertical bars**

**Left part of the figsize is the width and right side is the height**

```python
plt.figure(figsize=(7,3))
for group, data in dataset_loan.groupby('Property_Area'):
    plt.bar(group, data['ApplicantIncome'].mean())
    plt.xlabel('Property Area')
    plt.ylabel('Avg. Applicant Income')
    plt.title('Property Area vs. Avg. Applicant Income')
```

**Using the barh we are getting the horizontal bars**

```python
plt.figure(figsize=(7,3))
for group, data in dataset_loan.groupby('Property_Area'):
    plt.barh(group, data['ApplicantIncome'].mean())
    plt.xlabel('Avg. Applicant Income')
    plt.ylabel('Property Area')
    plt.title('Property Area vs. Avg. Applicant Income')
```

**There are two aspects 1) You can do the EDA first and then clean the data and the 2) we can clean the data and then do the EDA.** 

    1) EDA is first and cleaning later
        Benefits: We are analysing raw data and understanding what's going on
        Problem: If there are null values some of the charts will not get created
        
    2) Cleaning is first and EDA later
        Benefits: We can create any chart and analyse 
        Problem: We are manipulating the data and then we are doing the analysis

**Purpose of Boxplot**

    1) To understand the data distribution - The median line which is the middle black line the position of that line will tell us how the data is distributed. If the black line is at the lower side then the majority of the values are at the left side or at the lower side. However, if the majority of the values are at the higher side than the black line will be at the right side and if the data is normally distributed than the median will be at the center. Median is also known as Quartile 2 or 50 percentile. The median of the lower values or lower part is 25 percentile or it is also known as Quartile 1. The median of the upper values or upper part is also known as 75 percentile or Quartile 3. The difference between Quartile 3 and Quartile 1 is known as Inter Quartile Range (IQR). This IQR is nothing by middle 50% of the records
    
    2) To get the outliers - Python uses different formula however, the industry standard formula is any value less than Q1-1.5*IQR where 1.5 is a constant and any value more than Q3+1.5*IQR those are outliers

```python
sns.boxplot(data=dataset_car, y='Vehicle type', x='Fuel efficiency')
```
