**% Contribution calculation**

```python
tips.groupby('gender').agg({'total_bill':np.sum})
```

**It will generate the data vertically and give the average of female against the female records and same for the male**

```python
avg_sales = tips.groupby('gender').total_bill.transform('mean')
```

```python
avg_sales
```

**Saving that above variable in the new column in the same dataframe**

```python
tips['avg_sales'] = avg_sales
```

**Subtracting the total bill with average bill and storing in the new column diff**

```python
tips['diff'] = tips.total_bill - tips.avg_sales
```

Creating the new column cust_seg and updating the positve values as True and negative values as False**

```python
tips['Cust_seg'] = tips['diff']>0
```

**Updating the True to Above Average and False to Below Average using the map commmand**

```python
tips['Cust_seg'] = tips.Cust_seg.map({False:'Below Average', True:'Above Average'})
```

**Filtering all the values which are Below Average**

```python
tips[tips['Cust_seg']=='Below Average']
```

**Aggregation of gender and cust_seg to understand the total_bill average**

```python
tips.groupby(['gender','Cust_seg']).agg({'total_bill':np.mean})
```

**Sorting of the data based on the categorical data**

```python
df = pd.DataFrame({'orderid':[1,2,3,4,5,6],
             'Manufacturer':['A','A','B','C','D','E'],
             'shipmode':['Same Day', '1st Class', '2nd Class', 'Standard Class', 'Same Day','2nd Class']})
```

```python
df.sort_values('shipmode', ascending=False)
```

```python
from pandas.api.types import CategoricalDtype
rank_type = CategoricalDtype(['Standard Class', '2nd Class', '1st Class', 'Same Day'], ordered=True)
```

```python
rank_type
```

```python
df.info()
```

```python
df.shipmode
```

```python
df['shipmode'] = df.shipmode.astype(rank_type)
```

```python
df.shipmode
```

```python
df.sort_values('shipmode', ascending=False)
```

```python
df[df.shipmode>'2nd Class']
```
